```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
df1.total_income = label_encoder.fit_transform(df1.total_income)
```

```python
df1.union_name = label_encoder.fit_transform(df1.union_name)
```

```python
df1.gender = label_encoder.fit_transform(df1.gender)
```

```python
df1.diabetic = label_encoder.fit_transform(df1.diabetic)
```

```python
df1.profile_hypertensive = label_encoder.fit_transform(df1.profile_hypertensive)
```

```python
df1.RESULT_STAT_BP = label_encoder.fit_transform(df1.RESULT_STAT_BP)
```

```python
df1.RESULT_STAT_PR = label_encoder.fit_transform(df1.RESULT_STAT_PR)
```

```python
df1.head()
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
target1=ols('PULSE_RATE~total_income+union_name+gender+diabetic+profile_hypertensive+RESULT_STAT_BP+RESULT_STAT_PR',data=df1).fit()
```

```python
ans1=sm.stats.anova_lm(target1,type=2)
```

```python
ans1
```

```python
target2=ols('SYSTOLIC~total_income+union_name+gender+diabetic+profile_hypertensive+RESULT_STAT_BP+RESULT_STAT_PR',data=df1).fit()
```

```python
ans2=sm.stats.anova_lm(target2,type=2)
```

```python
ans2
```

```python
target3=ols('DIASTOLIC~total_income+union_name+gender+diabetic+profile_hypertensive+RESULT_STAT_BP+RESULT_STAT_PR',data=df1).fit()
```

```python
ans3=sm.stats.anova_lm(target3,type=2)
```

```python
ans3
```

# Analytic Outcomes
H0 = There is no significant difference

H1 = There is significant difference
1. The ”pulse rate” is significantly affected by whether someone is diabetic or not as null
hypothesis is rejected.
2. ”pulse rate” is significantly affected by the different ”union name” groups as alternative
hypothesis is accepted.
